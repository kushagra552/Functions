7. What are the advantages of using generators over regular functions?
 - Generators offer advantages over regular functions, particularly in memory efficiency and lazy evaluation, by yielding values one at a time instead of returning a complete result, making them ideal for handling large datasets or infinite sequences.
* Here's a more detailed breakdown:
 - Memory Efficiency:
Generators only compute and store the current value, not the entire sequence, which is crucial for handling large or infinite datasets that wouldn't fit in memory if generated all at once.
 - Lazy Evaluation:
Generators compute values only when they are needed, improving performance and responsiveness, especially in applications where the full sequence isn't required immediately.
 - Cooperative Multitasking:
Generators can pause execution and resume from where they left off, enabling cooperative multitasking, a feature not directly available in regular functions.
 - Readability and Conciseness:
Generators can lead to more readable and concise code, particularly when dealing with sequences or iterators.
 - Infinite Sequences:
Generators can be used to generate infinite sequences of values, which is impossible with regular functions that must return a finite result.

 - map(): Applies a given function to each item in an iterable and returns an iterator that yields the results. It transforms each element individually.

 - reduce(): Applies a function cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value. It requires importing from the functools module in Python 3.
